```python
# import BeautifulSoup
from bs4 import BeautifulSoup
import requests

# request icc-cricket.com for cricket teams ODI ranking webpage data
page1 = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page1
```

```python
# request icc-cricket.com for ODI batsmen ranking webpage data
page2 = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
page2
```

```python
# request icc-cricket.com for ODI bowlers ranking webpage data
page3 = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder')
page3
```

```python
# extract icc-cricket.com cricket teams ODI ranking webpage data
src_1 = BeautifulSoup(page1.content)
```

```python
# extract icc-cricket.com ODI batsmen ranking webpage data
src_2 = BeautifulSoup(page2.content)
```

```python
# extract icc-cricket.com ODI bowlers ranking webpage data
src_3 = BeautifulSoup(page3.content)
```

```python
print('List of Top 10 ODI Teams in women’s cricket:')

# scrape data for team rank
team_rank = []
for i in src_1.find_all("td",class_="rankings-block__banner--pos"):
    team_rank.append(i.text)
for i in src_1.find_all("td",class_="table-body__cell table-body__cell--position u-text-right"):
    team_rank.append(i.text)
team_rank = team_rank[:10] # cutting the list to only top 10

# scrape data for team name
team_name = []
for i in src_1.find_all("span",class_="u-hide-phablet"):
    team_name.append(i.text)
for i in src_1.find_all("span",class_="u-hide-phablet"):
    team_name.append(i.text)
team_name = team_name[:10] # cutting the list to only top 10

# scrape data for matches played by team
team_matches = []
for i in src_1.find_all('td',class_="rankings-block__banner--matches"):
    team_matches.append(i.text)
for i in src_1.find_all('td',class_="table-body__cell u-center-text"):
    team_matches.append(i.text)
team_matches = team_matches[:10] # cutting the list to only top 10

# scrape data for team points
team_points = []
for i in src_1.find_all('td',class_="rankings-block__banner--points"):
    team_points.append(i.text)
for i in src_1.find_all('td',class_="table-body__cell u-center-text"):
    team_points.append(i.text)
team_points = team_points[:10] # cutting the list to only top 10

# scrape data for team rating
team_rating = []
for i in src_1.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    team_rating.append(i.text.strip())
for i in src_1.find_all('td',class_="table-body__cell u-text-right rating"):
    team_rating.append(i.text)
team_rating = team_rating[:10] # cutting the list to only top 10

# display information collected on ODI team rankings (only top 10)
for val in range(0,len(team_rank)):
    print(team_rank[val],'|',team_name[val],'|',team_matches[val],'|',team_points[val],'|',team_rating[val])
    
print('\n****************\n')
print('List of Top 10 Women’s ODI Batting players:')

# scrape data for ODI batting players ranking
bat_rank = []
for i in src_2.find_all("span",class_="rankings-block__pos-number"):
    bat_rank.append(i.text.strip())
for i in src_2.find_all("span",class_="rankings-table__pos-number"):
    bat_rank.append(i.text.strip())
bat_rank = bat_rank[:10] # cutting the list to only top 10

# scrape data for ODI batting players name
bat_pl = []
for i in src_2.find_all('div',class_="rankings-block__banner--name-large"):
    bat_pl.append(i.text)
for i in src_2.find_all('td',class_="table-body__cell rankings-table__name name"):
    bat_pl.append(i.text.replace('\n',''))
bat_pl = bat_pl[:10] # cutting the list to only top 10

# scrape data for ODI batting players team
bat_team = []
for i in src_2.find_all('div',class_="rankings-block__banner--nationality"):
    bat_team.append(i.text.strip())
for i in src_2.find_all('span',class_="table-body__logo-text"):
    bat_team.append(i.text)
bat_team = bat_team[:10] # cutting the list to only top 10

# scrape data for ODI batting players rating
bat_rating = []
for i in src_2.find_all('div',class_="rankings-block__banner--rating"):
    bat_rating.append(i.text)
for i in src_2.find_all('td',class_="table-body__cell rating"):
    bat_rating.append(i.text)
bat_rating = bat_rating[:10] # cutting the list to only top 10

# display information collected on ODI batting players rankings (only top 10)
for val in range(0,len(bat_rank)):
    print(bat_rank[val],'|',bat_pl[val],'|',bat_team[val],'|',bat_rating[val])
    
print('\n****************\n')
print('List of Top 10 Women’s ODI all-rounder:')

# scrape data for women's cricket all-rounder player ranking
ar_rank = []
for i in src_3.find_all("span",class_="rankings-block__pos-number"):
    ar_rank.append(i.text.strip())
for i in src_3.find_all("span",class_="rankings-table__pos-number"):
    ar_rank.append(i.text.strip())
ar_rank = ar_rank[:10] # cutting the list to only top 10

# scrape data for women's cricket all-rounder player name
ar_pl = []
for i in src_3.find_all('div',class_="rankings-block__banner--name-large"):
    ar_pl.append(i.text)
for i in src_3.find_all('td',class_="table-body__cell rankings-table__name name"):
    ar_pl.append(i.text.replace('\n',''))
ar_pl = ar_pl[:10] # cutting the list to only top 10

# scrape data for women's cricket all-rounder player team
ar_team = []
for i in src_3.find_all('div',class_="rankings-block__banner--nationality"):
    ar_team.append(i.text.strip())
for i in src_3.find_all('span',class_="table-body__logo-text"):
    ar_team.append(i.text)
ar_team = ar_team[:10] # cutting the list to only top 10

# scrape data for women's cricket all-rounder player rating
ar_rating = []
for i in src_3.find_all('div',class_="rankings-block__banner--rating"):
    ar_rating.append(i.text)
for i in src_3.find_all('td',class_="table-body__cell rating"):
    ar_rating.append(i.text)
ar_rating = ar_rating[:10] # cutting the list to only top 10

# display information collected on women's cricket ODI all-rounder player rankings (only top 10)
for val in range(0,len(ar_rank)):
    print(ar_rank[val],'|',ar_pl[val],'|',ar_team[val],'|',ar_rating[val])
    
print('\n****************\n')
```

```python
# make dataframe to store the information on Top 10 ODI teams in women’s cricket
import pandas as pd

db_topteam = pd.DataFrame({'Rank':team_rank,'Team':team_name,'Matches':team_matches,'Points':team_points,'Rating':team_rating})
db_topteam.index +=1
db_topteam
```

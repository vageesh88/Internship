```python
# import BeautifulSoup
from bs4 import BeautifulSoup
import requests

# request wikipedia.org for webpage data
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
# extract wikipedia.org webpage data
src = BeautifulSoup(page.content)
src
```

```python
# scrape all header tags on page
header = []
for i in src.find_all("li",class_="mw-list-item"):
    header.append(i.text)

# display the list of header tags on Wikipedia homepage
print('List of Header Tags in Wikipedia Home page are:\n')
for i in range(0,len(header)):
    print(header[i])
```

```python
# import BeautifulSoup
from bs4 import BeautifulSoup
import requests

# request Google Scholar for webpage data
page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
page
```

```python
# extract data from Google Scholar
src = BeautifulSoup(page.content)
```

```python
# scrape data for rank of publication
rank = []
for i in src.find_all("td",class_="gsc_mvt_p"):
    rank.append(i.text.replace('\n',''))

# scrape data for publication name
publication = []
for i in src.find_all("td",class_="gsc_mvt_t"):
    publication.append(i.text.replace('\n',''))

# scrape data for h5 index
h5_index = []
for i in src.find_all("a",class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(i.text.replace('\n',''))

# scrape data for h5 median
h5_median = []
for i in src.find_all("span",class_="gs_ibl gsc_mp_anchor"):
    h5_median.append(i.text.replace('\n',''))

# dispaly data collected on top publication as per Google Scholar
for val in range(0,len(rank)):
    print(rank[val],publication[val],'|',h5_index[val],'|',h5_median[val])
```

```python
# make dataframe to store data on top publication information taken from Google Scholar
import pandas as pd

db = pd.DataFrame({'Rank':rank,'Publication':publication,'h5-index':h5_index,'h5-median':h5_median})
db.index +=1
db
```
